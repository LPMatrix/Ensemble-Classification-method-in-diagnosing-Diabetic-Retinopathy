```python
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedShuffleSplit, StratifiedKFold
from sklearn.preprocessing import Imputer
from xgboost import XGBRegressor
```

```python
#load dataset and separate dataset into features and labels
data = pd.read_csv('csv_result-messidor_features.csv')
#id column is useless, hence, removed
data.drop('id', axis=1, inplace = True)
data.head(5)
```

```python
X = data.drop('Class', axis = 1)
y = data['Class']
```

```python
# A parameter grid for XGBoost
params = {'min_child_weight':[4,5], 
          'nthread':[4],
          'gamma':[i/10.0 for i in range(3,6)],  
          'learning_rate': [.03, 0.05, .07, 0.1],
          'subsample':[i/10.0 for i in range(6,11)],
          'silent': [1],
          'colsample_bytree':[i/10.0 for i in range(6,11)], 
          'max_depth': [2,3,4],
          'n_estimators': [1000]}

my_XGB_model = XGBRegressor(nthread=-1,objective='reg:linear',missing=None,seed=8)
 
predictions_test_score = []


s_kf = StratifiedKFold(n_splits=4)
for train_index, test_index in s_kf.split(X, y):
    X_train = X.iloc[train_index]
    y_train = y.iloc[train_index]
    X_test = X.iloc[test_index]
    y_test = y.iloc[test_index]
    
    grid = RandomizedSearchCV(my_XGB_model, params, n_jobs=1,n_iter=10)
    grid.fit(X_train, y_train, verbose=False)
#     print(grid.best_params_)
    XGB_predictions = grid.predict(X_test)
    XGB_predictions = [round(value) for value in XGB_predictions]
    Accuracy = accuracy_score(y_test, XGB_predictions)
    predictions_test_score.append(Accuracy)
print("Accuracy:",predictions_test_score)
print("Average Accuracy:",np.mean(predictions_test_score))
```
